```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("allegations_20200726939.csv")
```

```python
df.head()
#mos = member of service
#fado = Force, Abuse of Authority, Discourtesy, Offensive Language
#PBBS = Patrol Borough Brooklyn South (New York Police Department)
#failure to give rtka card: NYPD did not provide personal information
```

```python
df.info()
```

```python
year = df.groupby(['year_received', 'complainant_gender']).count()
year[['complaint_id']]
```

```python
year_df = df.loc[(df['year_received'] > 2009) & (df['year_received'] < 2020)]
year_df.head()
```

```python
year = year_df.groupby('year_received').count()
year[['complaint_id']]
```

```python
sns.set(rc={'figure.figsize': (20,16)})
g = sns.countplot(x='year_received',hue='complainant_gender', palette='muted', data=year_df)
g = g.set(xlabel='Year of Complaint', ylabel = 'Number of Complaints')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
